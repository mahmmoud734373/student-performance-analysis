```python
plt.figure(figsize=(8,6))
plt.scatter(df["math score"], df["writing score"], alpha=0.6, color="blue")

plt.title("Scatter Plot: Math Score vs Writing Score")
plt.xlabel("Math Score")
plt.ylabel("Writing Score")
plt.grid(True)
plt.show()
```
